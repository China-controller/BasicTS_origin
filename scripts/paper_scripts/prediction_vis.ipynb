```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15.0, 3.0)
```

```python
from tqdm import tqdm
import random
```

```python
def plot(prediction, real_value, history, batch_id, node_id, plot_res=True, save=True, line_width=2, font_size=24, num_xticks=6):
    # plot data
    prediction_data = prediction[batch_id, :, node_id, 0]
    real_value_data = real_value[batch_id, :, node_id, 0]
    history_data = history[batch_id, :, node_id, 0]
    residual_data = prediction_data - real_value_data
    ## plot data
    plt.plot(range(history_data.shape[0]), history_data, label='History', linewidth=line_width, color='black')
    plt.plot(range(history_data.shape[0], history_data.shape[0]+prediction_data.shape[0]), prediction_data, label='Prediction', linewidth=line_width, color="green")
    plt.plot(range(history_data.shape[0], history_data.shape[0]+real_value_data.shape[0]), real_value_data, label='Real Value', linewidth=line_width, color="red")
    if plot_res:
        plt.plot(range(history_data.shape[0], history_data.shape[0]+residual_data.shape[0]), residual_data, label='Residual', linewidth=line_width)
    plt.grid()
    plt.legend(prop={'size': font_size})
    xticks = range(num_xticks)
    plt.xticks([_*int(prediction_data.shape[0]/num_xticks) for _ in xticks], size=font_size, rotation=0)
    plt.yticks(size=font_size)
    if save:
        plt.savefig("prediction_vis.png") 
    plt.show()
    plt.clf()
    
```

```python
MODEL_NAME = 'DLinear'
DATASET_NAME = 'ETTh1'
with open("inference_results_{0}_{1}.pkl".format(MODEL_NAME, DATASET_NAME), "rb") as f:
    inference_results = pickle.load(f)
prediction_DLinear = inference_results["prediction"].cpu().numpy()
real_value = inference_results["real_value"].cpu().numpy()
history = inference_results["history"].cpu().numpy()

MODEL_NAME = 'Autoformer'
DATASET_NAME = 'ETTh1'
with open("inference_results_{0}_{1}.pkl".format(MODEL_NAME, DATASET_NAME), "rb") as f:
    inference_results = pickle.load(f)
prediction_Autoformer = inference_results["prediction"].cpu().numpy()
real_value = inference_results["real_value"].cpu().numpy()
history = inference_results["history"].cpu().numpy()

MODEL_NAME = 'FEDformer'
DATASET_NAME = 'ETTh1'
with open("inference_results_{0}_{1}.pkl".format(MODEL_NAME, DATASET_NAME), "rb") as f:
    inference_results = pickle.load(f)
prediction_FEDformer = inference_results["prediction"].cpu().numpy()
real_value = inference_results["real_value"].cpu().numpy()
history = inference_results["history"].cpu().numpy()

MODEL_NAME = 'Crossformer'
DATASET_NAME = 'ETTh1'
with open("inference_results_{0}_{1}.pkl".format(MODEL_NAME, DATASET_NAME), "rb") as f:
    inference_results = pickle.load(f)
prediction_Crossformer = inference_results["prediction"].cpu().numpy()
real_value = inference_results["real_value"].cpu().numpy()
history = inference_results["history"].cpu().numpy()

MODEL_NAME = 'PatchTST'
DATASET_NAME = 'ETTh1'
with open("inference_results_{0}_{1}.pkl".format(MODEL_NAME, DATASET_NAME), "rb") as f:
    inference_results = pickle.load(f)
prediction_PatchTST = inference_results["prediction"].cpu().numpy()
real_value = inference_results["real_value"].cpu().numpy()
history = inference_results["history"].cpu().numpy()
```

```python
batch_size, _, variate_num, feature_num = history.shape

# randomly select batch_id (0~batch_size) and variate_id (0~variate_num)
batch_id = random.randint(0, batch_size-1)
variate_id = random.randint(0, variate_num-1)
```

```python
# plot data
prediction_data_Autoformer = prediction_Autoformer[batch_id, :, variate_id, 0]
prediction_data_FEDformer = prediction_FEDformer[batch_id, :, variate_id, 0]
prediction_data_Crossformer = prediction_Crossformer[batch_id, :, variate_id, 0]
prediction_data_PatchTST = prediction_PatchTST[batch_id, :, variate_id, 0]
real_value_data = real_value[batch_id, :, variate_id, 0]
history_data = history[batch_id, :, variate_id, 0]
## plot data
plt.plot(range(history_data.shape[0]), history_data, label='History', linewidth=2, color='black')
plt.plot(range(history_data.shape[0], history_data.shape[0]+real_value_data.shape[0]), real_value_data, label='Real Values', linewidth=1.5, color="green", linestyle='--')
plt.plot(range(history_data.shape[0], history_data.shape[0]+prediction_data_Autoformer.shape[0]), prediction_data_Autoformer, label='Autoformer', linewidth=2, color="red")
plt.grid()
plt.legend(prop={'size': 16})
plt.xticks(size=16)
plt.yticks(size=16)
plt.savefig("prediction_vis_Autoformer.png", bbox_inches='tight')
plt.show()
plt.clf()
```
